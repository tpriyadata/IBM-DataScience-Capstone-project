
### Task 2


#####  Display 5 records where launch sites begin with the string 'KSC' 


```python
pd.read_sql("SELECT * FROM SPACEXTABLE WHERE Launch_Site LIKE 'KSC%' LIMIT 5", con)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
pd.read_sql("""SELECT Customer, SUM(PAYLOAD_MASS__KG_) as Total_Payload_Mass 
               FROM SPACEXTABLE 
               WHERE Customer = 'NASA (CRS)'""", con)
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
pd.read_sql("""SELECT Booster_Version, AVG(PAYLOAD_MASS__KG_) as Avg_Payload_Mass 
               FROM SPACEXTABLE 
               WHERE Booster_Version LIKE 'F9 v1.1%'""", con)
```

### Task 5

##### List the date where the succesful landing outcome in drone ship was acheived.


_Hint:Use min function_ 


```python
pd.read_sql("""SELECT MIN(Date) as First_Successful_Landing 
               FROM SPACEXTABLE 
               WHERE Landing_Outcome = 'Success (ground pad)'""", con)
```

### Task 6

##### List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
pd.read_sql("""SELECT Booster_Version, PAYLOAD_MASS__KG_, Landing_Outcome 
               FROM SPACEXTABLE 
               WHERE Landing_Outcome = 'Success (drone ship)' 
               AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000""", con)
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
pd.read_sql("""SELECT Mission_Outcome, COUNT(*) as Count 
               FROM SPACEXTABLE 
               GROUP BY Mission_Outcome 
               ORDER BY Count DESC""", con)
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass. Use a subquery.


```python
pd.read_sql("""SELECT Booster_Version, PAYLOAD_MASS__KG_ 
               FROM SPACEXTABLE 
               WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTABLE)""", con)
```

### Task 9


##### List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017
**Note: SQLLite does not support monthnames. So you need to use substr(Date,6,2) for month, substr(Date,9,2) for date, substr(Date,0,5),='2017' for year.**


```python
pd.read_sql("""SELECT Date, Booster_Version, Launch_Site, Landing_Outcome 
               FROM SPACEXTABLE 
               WHERE Landing_Outcome LIKE 'Failure (drone ship)%' 
               AND Date LIKE '2015%'""", con)
```

```python
pd.read_sql("""
    SELECT 
        CASE substr(Date, 6, 2)
            WHEN '01' THEN 'January'
            WHEN '02' THEN 'February'
            WHEN '03' THEN 'March'
            WHEN '04' THEN 'April'
            WHEN '05' THEN 'May'
            WHEN '06' THEN 'June'
            WHEN '07' THEN 'July'
            WHEN '08' THEN 'August'
            WHEN '09' THEN 'September'
            WHEN '10' THEN 'October'
            WHEN '11' THEN 'November'
            WHEN '12' THEN 'December'
        END AS Month_Name,
        Booster_Version,
        Launch_Site,
        Landing_Outcome
    FROM SPACEXTABLE
    WHERE Landing_Outcome = 'Success (ground pad)'
    AND substr(Date, 0, 5) = '2017'
    ORDER BY substr(Date, 6, 2)
""", con)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
pd.read_sql("""
    SELECT 
        Landing_Outcome,
        COUNT(*) as Count
    FROM SPACEXTABLE
    WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
    GROUP BY Landing_Outcome
    ORDER BY Count DESC
""", con)
```
